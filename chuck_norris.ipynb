```python
navigator='Mozilla/5.0(Macintosh;Intel Mac OS X 10_10_1)AppleWebKit/537.36(KHTML, like Gecko) Chrome/41.0.2227'

url ="https://chucknorrisfacts.net/facts.php?page=3"

html= requests.get(url,headers={'User-Agent':navigator})
soup= BeautifulSoup(html.text,'html.parser')

factbody = soup.find_all("div",{"style":"border-top: 1px solid #251836; padding: 0 0 5px 7px;"})

for i in factbody :
    print(i.get_text().split("\n")[1].split("Rated")[0])


```

```python
### récupération des 5 premières pages  grace à un traitement automatisé ()
```

```python
## pour récupérer tous les éléments on fait une boucle imbriqué ds celle là 
##

navigator='Mozilla/5.0(Macintosh;Intel Mac OS X 10_10_1)AppleWebKit/537.36(KHTML, like Gecko) Chrome/41.0.2227'

for num in range(1,5) :
    url ="https://chucknorrisfacts.net/facts.php?page="+ str(num)
    html= requests.get(url,headers={'User-Agent':navigator})
    soup= BeautifulSoup(html.text,'html.parser')

    factbody = soup.find_all("div",{"style":"border-top: 1px solid #251836; padding: 0 0 5px 7px;"})

    for i in factbody :
        print(i.get_text().split("\n")[1].split("Rated")[0])
```
